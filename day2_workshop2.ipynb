```python
import pandas as pd   #library
df=pd.read_csv("outbreak_detect.csv")
df
```

```python
#visualizing _NAN values-NULL in the dataset
import matplotlib.pyplot as plt
import seaborn as sns
def get_heatmap(df):
    plt.figure(figsize=(6,4)) #width,breadth
    sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
    plt.tight_layout()
    return plt.show()
get_heatmap(df)
```

```python
#deleting the null value rows from dataset
df=df.dropna()
df
```

```python
#visualizing _NAN values-NULL in the dataset
import matplotlib.pyplot as plt
import seaborn as sns
def get_heatmap(df):
    plt.figure(figsize=(6,4)) #width,breadth
    sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
    plt.tight_layout()
    return plt.show()
get_heatmap(df)
```

```python
from sklearn import preprocessing
LE=preprocessing.LabelEncoder()
df.Outbreak=LE.fit_transform(df.Outbreak)
print(df.head())
```

```python
#removing some unwanted 
df=df.drop('Positive',axis=1)
df=df.drop('pf',axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sst=StandardScaler()
```

```python
data_scaled=df.iloc[:,:-1].values

data_scaled=sst.fit_transform(data_scaled)
data_scaled=pd.DataFrame(data_scaled)
data_scaled.head()
```

```python
data_scaled.columns='minTemp','maxTemp','avgHumidity','Rainfall'
```

```python
data_scaled.head()
```

```python
data_scaled['class']=df.Outbreak
data_scaled
```

```python
data_scaled=data_scaled[data_scaled['class'].notna()]
x=data_scaled.iloc[:,0:4]
y=data_scaled.iloc[:,4:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=3)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```
